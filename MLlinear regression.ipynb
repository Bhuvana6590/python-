```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from word2number import w2n
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#import Dataset
df=pd.read_csv('hiring.csv')
df.head()
```

```python
#preprocessing dataset
df.experience=df.experience.fillna('zero')
df
```

```python
#Apply word to number
df.experience=df.experience.apply(w2n.word_to_num)
df
```

```python
#floor value to convert point value to whole num
import math
median_test_score=math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
dff=df['test_score(out of 10)'].mean()
dff
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(dff)
df
```

```python
#show column name
df.columns
```

```python
#Feature selection
predictors=['experience', 'test_score(out of 10)','interview_score(out of 10)']
x=df[predictors]
y=df['salary($)']
```

```python
#Split train and test datasets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
